## Fixing header markups for Myanmar fonts
Header markups lenght is equal or greater than number of character in Header Text.

Myanmar font make difficult to make correct length of header markups.

solution: find existing header markups and read and count length of Myanmar text above it and adjust the length of header markups respectively


```python
# fixing header markups for Myanmar texts
import re

def fixHeaderMarkups(file_path):
    
    out_file_path = file_path[:-4]+"_fixed.rst"
    
    header_patterns = re.compile(r'(^=+$)|(^-+$)|(^\.+$)|(^\^+$)')  #(r'^(=+)(.*?)\1$', re.MULTILINE)  # Matches section headers
    
    # Initialize a list to store line numbers with matches
    header_line_numbers = []

    # Open and read the file
    with open(file_path, 'r', encoding="utf8") as file:
        txt_lines = file.readlines()
        for line_number, line in enumerate(txt_lines):
            # Use re.search() to search for the pattern in each line
            if re.search(header_patterns, line):
                header_line_numbers.append(line_number)

    # fix header markups
    for line_id in header_line_numbers:
        markup = txt_lines[line_id][0]
        header_txt = txt_lines[line_id-1]
        if len(header_txt):
            markup = "".ljust(len(header_txt),markup)
            txt_lines[line_id] = markup+"\n"
            print (f"line #: {line_id} is fixed") 
    
    # write fixed file
    with open(out_file_path, 'w', encoding="utf8") as file:
        file.writelines(txt_lines)
        
    print (f"Header markups are fixed and save as {out_file_path}")
```

# Fixing table row and cell markups for table with Myanmar Texts

solution: scan lines with row and cell markups. find longest length of text in each column. adjust

```python
file = r"C:\Users\Kyaw Naing Win\Downloads\georeferencer_temp.rst"

def findTables(file):
    # first border markups line
    # last border markups line
    # get length of each cells
    # get leght of text in each cell
    # border markups line
    tbl_border_pattern = re.compile('(^\+.*\+$)')
    tbl_rows = []
    
    with open(file,'r',encoding='utf8') as f:
        txt_lines = f.readlines()
    for index,line in enumerate(txt_lines):
        if len(tbl_border_pattern.findall(line)):
            tbl_rows.append(index)
            
    tbls= [False]
    tbls = tbls + ([(x1-x0)>2 for x0,x1 in zip(tbl_rows[:-1],tbl_rows[1:])])
    tbls.append(True)
    
    start_rows=[tbl_rows[0]]
    end_rows=[]

    for i,t in enumerate(tbls[:-1]):
        #print (t,not(tbls[i+1]))
        try:
            if t and not(tbls[i+1]):
                end_rows.append(tbl_rows[i-1])
                continue
            if not(t) and tbls[i+1]:
                start_rows.append(tbl_rows[i+1])
                continue
        except:
            # last row
            end_rows.append(tbl_rows[i])
    
    return start_rows,end_rows
```

```python
tbl_start_lines,tbl_end_lines = findTables(file)
tbl_start_lines,tbl_end_lines
```

```python
#scanning column widths by border markups
tbl_col_widths = txt_lines[tbl_start_lines[0]].split("+")
tbl_col_widths = [len(cell) for cell in tbl_col_widths]

#scanning column widths by row markups
# exclude "|" that are part of cell text
tbl_data_widths = []
for i in range(tbl_start_lines[0]+1,tbl_end_lines[0],2):
    line = txt_lines[i]
    pattern = r'\|([^\s|]+)\|'
    line = re.sub(pattern, r'@\1@', line)    
    temp_data = line.split("|")
    tbl_data_widths.append([len(data) for data in temp_data])
```

```python
# Converting to csv format table to sovle the myanmar unicode issues
'''
.. csv-table:: Example :rst:dir:`csv-table`
   :header: "Header Text 1", "Header Text 2", "Header Text 3"

   "Data 1.1",               "Data 1.2",      "Data 1.3"
   "Data 2.1",               "Data 2.2",      "Data 2.3"
   "Data 3.1",               ,                "Data 1.3 (Data 1.2 left out)"
   "Data 4.1",               "Data 4.2 (Data 4.3 left out)"
'''
with open(file,'r',encoding='utf8') as f:
        txt_lines = f.readlines()

#for j,cell_widths in enumerate(tbl_data_widths):
for j in range(tbl_start_lines[0]+1,tbl_end_lines[0],2):
    line = txt_lines[j] # only data rows are checked
    #print(line)
    pattern = r'\|([^\s|]+)\|'
    line = re.sub(pattern, r'@\1@', line)
    line = line.split("|")
    for i,cell in enumerate(line):
        cell = cell.replace("@","|")
        cell = cell.strip()
        cell = '\"'+ cell +'\"'
        line[i] = cell

    txt_lines[j] = ",".join(line)
    
```
